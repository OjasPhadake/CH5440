```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import date
```

# Question 2

Now, I am defining a function which is titled "LinearRegression" and which carries out OLS and plots the graph of the data points and the linearly regressed line. At the end, it returns the slope and intercept of the thus plotted line as well. 

```python
def LinearRegression(x, y, xtitle, ytitle, title):
    xaxis = np.linspace(np.min(x),np.max(x)+1,2*len(x))
    xmean = np.mean(x)
    ymean = np.mean(y)

    S_xy = np.dot(x-xmean, y-ymean)
    S_xx = np.dot(x-xmean, x-xmean)
    
    slope = S_xy/S_xx
    intercept = ymean - slope*xmean
    plt.scatter(x, y, c='r', marker='.', label="Measured Data")
    plt.plot(xaxis, slope*xaxis + intercept, label="OLS")
    plt.xlabel(xtitle)
    plt.ylabel(ytitle)
    plt.legend()
    plt.grid(True)
    plt.title(title)

    return slope, intercept
```

```python
df = pd.read_csv("epa-sea-level.csv")
df.head()
df = df.dropna()

col1 = df[df.columns[0]]
col2 = df[df.columns[1]]
```

```python
year_arr = []
for i in range(len(col1)):
    yr = pd.Timestamp(col1[i]).year
    year_arr.append(yr)
plt.plot(year_arr, col2)
plt.xlabel("Year")
plt.ylabel("(mm/year) increase in height")
plt.title("Increase of water per year with all data")
```

Now plotting a graph which will include the height which has increased against the levels existing in the year 1880. 

```python
prefix = np.zeros(len(col2))
for i in range(1, len(col2)):
    prefix[i] = prefix[i-1] + col2[i]

plt.scatter(year_arr, prefix, marker='.')
plt.xlabel("Year")
plt.ylabel("Height increase over 1880 (mm)")
plt.title("Height increase over 1880 vs Year")
```

#### Considering the entire dataset, we have:

```python
slope, intercept = LinearRegression(year_arr, col2, "Year", "(mm/year) increase in height", "Increase of water per year with all data")
```

```python
recent_yr = year_arr[len(year_arr)-50:]
recent_hts = col2[len(year_arr)-50:]
```

### Considering only the recent 50 year data, we get:

```python
slope2, intercept2 = LinearRegression(recent_yr, recent_hts, "Year", "(mm/year) increase in height", "Increase of water per year with recent 50 year data")
```

```python
h = 750
rise=0
cnt=2014
while rise<=750:
    rise+=(cnt*slope+intercept)
    cnt+=1
print(f"Considering the entire dataset, the level of water will be 75cm from current levels along the Chennai coastline will be reached in {cnt-2014} years.\nThen, the land inside 1km along the coast will come under water. \nThe year is {cnt} when it'll be reached. ")
```
