The function defined below carries out Ordinary Linear Regression, and also does the plotting of the data points and the regressed line. 

```python
def LinearRegression(x, y, xtitle, ytitle, title):
    xaxis = np.linspace(np.min(x),np.max(x)+1,len(x))
    xmean = np.mean(x)
    ymean = np.mean(y)

    S_xy = np.dot(x-xmean, y-ymean)
    S_xx = np.dot(x-xmean, x-xmean)
    
    slope = S_xy/S_xx
    intercept = ymean - slope*xmean
    
    plt.scatter(x, y, c='r', marker='.', label="Measured Data")
    plt.plot(xaxis, slope*xaxis + intercept, label="OLS")
    plt.xlabel(xtitle)
    plt.ylabel(ytitle)
    plt.legend()
    plt.grid(True)
    plt.title(title)

    return slope, intercept
```

The function defined below carries out Total Linear Regression, and also does the plotting of the data points and the regressed line. 

```python
def TotalLinearRegression(x, y, xtitle, ytitle, title):
    xaxis = np.linspace(np.min(x),np.max(x)+1,2*len(x))
    xmean = np.mean(x)
    ymean = np.mean(y)
    rho = np.square(np.std(x)/np.std(y))

    S_xy = np.dot(x-xmean, y-ymean)
    S_xx = np.dot(x-xmean, x-xmean)
    S_yy = np.dot(y-ymean, y-ymean)

    
    slope = ((S_yy - rho*S_xx) + np.sqrt((S_yy - rho*S_xx)**2 + 4*rho*(S_xy**2)))/(2*S_xy)
    intercept = ymean - slope*xmean
    plt.scatter(x, y, c='r', marker='.', label="Measured Data")
    plt.plot(xaxis, slope*xaxis + intercept, label="TLS")
    plt.xlabel(xtitle)
    plt.ylabel(ytitle)
    plt.legend()
    plt.grid(True)
    plt.title(title)

    return slope, intercept
```

### Solution:

```python
slope1, intercept1 = LinearRegression(yr, CO2,"Year", "CO2 Concentrations (ppm)",  "CO2 Concentration vs Year (OLS)")
```

```python
slope2, intercept2 = LinearRegression(yr, temp, "Year", "Temperature (°F)", "Temperature vs Year (OLS)")
```

```python
deg_max = 3.6
yr_reach1 = (deg_max - intercept2)/slope2
CO2_max1 = yr_reach1*slope1 + intercept1
print('{0:.4f}'.format(CO2_max1) , "ppm is the highest tolerable concentration. ")
print('{0:.0f}'.format(yr_reach1), "is the year in which the maximum permissible level of CO2 in the atmosphere will be reached.")
```

TLS is carried out in the next cell between Temperature and CO2 Concentration. 

```python
slope3, intercept3 = TotalLinearRegression(CO2, temp, "CO2 Concentrations (ppm)", "Temperature (°F)", "Temperature vs CO2 Concentration (TLS)")
```

```python
deg_max = 3.6
CO2_max2 = (deg_max - intercept3)/slope3
yr_reach2 = (CO2_max2 - intercept1)/slope1
print('{0:.4f}'.format(CO2_max2), "ppm is the highest possible concentration possible as per TLS.")
print('{0:.0f}'.format(yr_reach2), "is the year in which the maximum permissible level of CO2 in the atmosphere will be reached according to TLS. ")
```
